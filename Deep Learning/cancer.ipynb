```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\Orhan Bayrak\Desktop\Yazilim\Python\ML Projects\Datasets\Cancer_Data.csv")
data.head(3)
```

```python
data.drop(columns=["id","Unnamed: 32"],axis=1,inplace=True)
data.head(3)
```

```python
data["diagnosis"]=data["diagnosis"].map({
    "M":1,
    "B":0
})
data.head(3)
```

```python
data.corr()
```

```python
M=data[data["diagnosis"]==1]
B=data[data["diagnosis"]==0]

plt.scatter(M.radius_mean,M.texture_mean,color="red",alpha=0.5)
plt.scatter(B.radius_mean,B.texture_mean,color="green",alpha=0.5)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X=data.drop(columns=["diagnosis"],axis=1)
y=data["diagnosis"]

X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)

sc=StandardScaler()
X_train_sc=sc.fit_transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input

model=tf.keras.models.Sequential([
    Input(shape=(X_train_sc.shape[1],)),
    tf.keras.layers.Dense(units=16,activation="relu"),
    tf.keras.layers.Dense(units=32,activation="relu"),
    tf.keras.layers.Dense(units=1,activation="sigmoid")
])

model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
model.fit(X_train_sc,y_train,epochs=100)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = (model.predict(X_test_sc) > 0.5)

cm=confusion_matrix(y_test,y_pred)
cm
```
