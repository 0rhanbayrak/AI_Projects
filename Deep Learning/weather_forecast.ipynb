Explanation:Goal of this project is to make future predictions on a time serie by using LSTM.

Method:I used the weather data obtained between 2021-01-01 and 2023-01-31 in Istanbul.I had the model predict the January data of 2023 and compared the results with the actual data.

Libraries:pandas,numpy,plotly,tensorflow

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r"FILE PATH => istanbul-weather-data")
data.tail(3)
```

```python
data.drop(columns=["prcp","snow","wpgt","tsun"],inplace=True)
data.head(3)
```

```python
data["date"] = pd.to_datetime(data["date"])
data.set_index("date", inplace=True) 
data.head() 
```

```python
data.isnull().sum()
```

```python
import plotly.graph_objects as go

figure = go.Figure(data=[go.Scatter(x=data.index, y=data["tavg"], mode='lines', name='Avg Temp')])

figure.update_layout(title="Average Temperature",
                     xaxis_title="Date",
                     yaxis_title="Temperature (°C)",
                     xaxis_rangeslider_visible=False)

figure.show()
```

```python
data.corr()
```

```python
data.head(1)
```

```python
train_data=data.loc["2021-01-01":"2022-12-31"]
test_data=data.loc["2023-01-01":"2023-01-31"]

X_train=train_data[["tmin","tmax","wdir","wspd","pres"]].to_numpy()
y_train=train_data["tavg"].to_numpy().reshape(-1,1)

X_test=test_data[["tmin","tmax","wdir","wspd","pres"]].to_numpy()
y_test=test_data["tavg"].to_numpy().reshape(-1,1)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input

model = tf.keras.models.Sequential([
    Input(shape=(X_train.shape[1], 1)),
    tf.keras.layers.LSTM(128, return_sequences=True),  
    tf.keras.layers.LSTM(64, return_sequences=False),  
    tf.keras.layers.Dense(25),  
    tf.keras.layers.Dense(1)  
])

model.compile(optimizer="adam",loss="mean_squared_error")

model.fit(X_train.reshape(X_train.shape[0], X_train.shape[1], 1), y_train, epochs=30, batch_size=1)
```

```python
X_test_reshaped = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

predictions = model.predict(X_test_reshaped)

predicted_temperatures = pd.DataFrame(predictions, index=test_data.index, columns=["Predicted Temperature"])

import plotly.graph_objects as go

figure = go.Figure([
    go.Scatter(x=test_data.index, y=test_data["tavg"], mode='lines', name='Real Temp'),
    go.Scatter(x=predicted_temperatures.index, y=predicted_temperatures["Predicted Temperature"], mode='lines', name='Predicted Temp', line=dict(dash='dash'))
])

figure.update_layout(title="Temperature Prediction for January 2023",
                     xaxis_title="Date",
                     yaxis_title="Temperature (°C)",
                     xaxis_rangeslider_visible=False)

figure.show()
```
