Explanation:Goal of this project is to find the correlations between movie audiences.Results can be used to make advices to users with some Machine Learning algorithms.

Libraries:pandas,numpy

```python
import pandas as pd
import numpy as np
data=pd.read_csv(r"FILE PATH => users.data.csv",sep="\t",names=["user_id","item_id","rating","timestamp"])
data.head()  
```

```python
movies=pd.read_csv(r"FILE PATH => movie_id_titles.csv")
movies.head()
```

```python
new_data=pd.merge(data,movies,on="item_id")
new_data.head()
```

```python
pivot_table=pd.pivot_table(data=new_data,index="user_id",columns="title",values="rating")
pivot_table.head()
```

```python
starwars=pivot_table["Star Wars (1977)"]
starwars.head()
```

```python
strw_correlation=pivot_table.corrwith(starwars)
strw_correlation.head()
```

```python
strw_correlation=pd.DataFrame(strw_correlation,columns=["Correlation"])
strw_correlation.dropna(inplace=True)
strw_correlation.head()
```

```python
strw_correlation.sort_values("Correlation",ascending=False)
```

```python
data.drop(columns="timestamp",inplace=True) 
data.head()
```

```python
ratings=pd.DataFrame(new_data.groupby("title")["rating"].mean())
ratings.sort_values("rating",ascending=False)
```

```python
ratings["Vote Count"]=pd.DataFrame(new_data.groupby("title")["rating"].count())
ratings.head(5)
```

```python
ratings.sort_values("Vote Count",ascending=False)
```

```python
strw_correlation["Vote Count"]=ratings["Vote Count"]
strw_correlation
```

```python
strw_correlation[strw_correlation["Vote Count"]>100].sort_values("Correlation",ascending=False).head(10)
```
